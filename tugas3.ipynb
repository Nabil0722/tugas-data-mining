```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
file_path = '/content/Dataset1.csv'
dataset = pd.read_csv(file_path, delimiter=';')
```

```python
print(dataset.head())
```

```python
le = LabelEncoder()
```

```python
dataset['Device Model'] = le.fit_transform(dataset['Device Model'])
dataset['Operating System'] = le.fit_transform(dataset['Operating System'])
dataset['Gender'] = le.fit_transform(dataset['Gender'])
```

```python
scaler = StandardScaler()
```

```python
numerical_columns = ['Battery Drain (mAh/day)', 'Age']
dataset[numerical_columns] = scaler.fit_transform(dataset[numerical_columns])
```

```python
print("Ringkasan Statistik:")
print(dataset.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(dataset['Battery Drain (mAh/day)'], bins=20, kde=True)
plt.title('Distribusi Battery Drain (mAh/day)')
plt.xlabel('Battery Drain (mAh/day)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x='Age', y='Battery Drain (mAh/day)', hue='Gender')
plt.title('Hubungan antara Usia dan Battery Drain')
plt.xlabel('Usia')
plt.ylabel('Battery Drain (mAh/day)')
plt.legend()
plt.show()
```
